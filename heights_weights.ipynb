```python
import os.path
import math
import matplotlib.pyplot as plt
import numpy as np

plt.close("all")
plt.rcParams['xtick.major.pad']='8'
plt.rcParams['ytick.major.pad']='8'

### CHANGE PATH TO WHEREVER YOU SAVE THE DATA FILES ### 
data = np.loadtxt("/content/heights_weights.txt") 

n_train = 20000
n_test = 1000

# Randomly select training, test and validation sets
np.random.shuffle(data)
data_train = data[range(n_train),:]
data_test = data[range(n_train, n_train+n_test),:]
heights_train = data_train[:,1]
weights_train = data_train[:,2]
heights_test = data_test[:,1]
weights_test = data_test[:,2]



### INSERT CODE HERE ###
# Hint: To fit model 1 use the expression you found directly
# to fit model 2 you can use np.linalg.lstsq

num = 0
den = 0
for x,y in zip(heights_train, weights_train):
  num += x*y
  den += x**2

pred_model_1 = num/den

A = np.vstack([heights_train, np.ones(len(heights_train))]).T
pred_model_2 = np.linalg.lstsq(A, weights_train, rcond=None)[0]






plt.figure(figsize=(10, 7))
plt.title('Model 1 - No Intercept')
plt.plot(heights_test, weights_test, '.',  color='skyblue',markeredgecolor='blue')
plt.plot(heights_test, pred_model_1*heights_test,  c="red")

plt.figure(figsize=(10, 7))
plt.title('Model 2 - With Intercept')
plt.plot(heights_test, weights_test, '.', color='skyblue',markeredgecolor='blue')
plt.plot(heights_test, heights_test*pred_model_2[0] + pred_model_2[1], c="red")

error_1 = np.sum(np.abs((pred_model_1*heights_test  - weights_test)/weights_test))/n_test
error_2 = np.sum(np.abs(((heights_test*pred_model_2[0] + pred_model_2[1]) - weights_test)/weights_test))/n_test

print("Model 1 relative error: " + str(error_1))
print("Model 2 relative error: " + str(error_2))
```
